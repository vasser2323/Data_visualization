### Панели

```python
chart_point1 = alt.Chart(cars).mark_point().encode(
    x='Horsepower',
    y='Miles_per_Gallon',
    color='Origin'
).properties(
    title='Horsepower vs. Miles per Gallon'
)

chart_line1 = alt.Chart(cars).mark_line().encode(
    x='Year',
    y='mean(Miles_per_Gallon)',
    color='Origin'
).properties(
    title='Average MPG over Years'
)

chart_tick1 = alt.Chart(cars).mark_tick().encode(
    x='Cylinders:O',
    y='Miles_per_Gallon',
    color='Origin'
).properties(
    title='MPG by Number of Cylinders'
)

chart_bar1 = alt.Chart(cars).mark_bar().encode(
    x='mean(Horsepower):Q',
    y='Origin:N',
    color='Origin'
).properties(
    title='Average Horsepower by Origin'
)
```

```python
combined_chart1 = (chart_point1 | chart_line1) & (chart_tick1 | chart_bar1)
combined_chart1.show()
```

```python
chart_area1 = alt.Chart(cars).mark_area().encode(
    x='Year:T',
    y='sum(Weight_in_lbs):Q',
    color='Origin:N'
).properties(
    title='Total Car Weight over Years'
)

chart_rect1 = alt.Chart(cars).mark_rect().encode(
    x='Horsepower:Q',
    y='Miles_per_Gallon:Q',
    color='count()'
).properties(
    title='Density of Cars by Horsepower and MPG'
)

chart_shape1 = alt.Chart(cars).mark_point().encode(
    x='Displacement',
    y='Acceleration',
    shape='Origin',
    color='Origin'
).properties(
    title='Acceleration vs. Displacement by Origin'
)

chart_size1 = alt.Chart(cars).mark_point().encode(
    x='Horsepower',
    y='Miles_per_Gallon',
    size='Weight_in_lbs',
    color='Origin'
).properties(
    title='Horsepower vs. MPG with Size by Weight'
)


combined_chart2 = (chart_area1 | chart_rect1) & (chart_shape1 | chart_size1)
combined_chart2.show()
```

1. Используя данные cars, создайте диаграмму рассеяния, где размер точек становится больше при наведении.

2. Используя данные cars, создайте двухпанельный график с гистограммами (например, миль на галлон на одной панели, мощность (horsepower) – на другой), в котором выделение данных на одной гистограмме фильтрует данные на второй.

### Диаграмма рассеяния с увеличением размера точек при наведении

```python
hover = alt.selection_point(on='mouseover')

scatter_hover = alt.Chart(cars).mark_point().encode(
    x='Horsepower',
    y='Miles_per_Gallon',
    color='Origin',
    size=alt.condition(hover, alt.value(200), alt.value(50))
).add_selection(
    hover
).properties(
    title='Scatter Plot with Hover Effect'
)
scatter_hover.show()
```

### Двухпанельный график с гистограммами и выделением

```python
brush = alt.selection_interval()

hist_mpg=alt.Chart(cars).mark_bar().encode(
    x=alt.X('Miles_per_Gallon:Q', bin=True),
    y='count()',
    color=alt.condition(brush, 'Origin:N', alt.value('lightgray'))
).add_selection(brush).properties(
    title='Histogram of MPG'
)

hist_horsepower = alt.Chart(cars).mark_bar().encode(
    x=alt.X('Horsepower', bin=True),
    y='count()',
    color=alt.condition(brush, 'Origin:N', alt.value('lightgray'))
).transform_filter(
    brush
).properties(
    title='Histogram of Horsepower'
)

linked_histograms = hist_mpg | hist_horsepower
linked_histograms.show()
```

```python
import pandas as pd
import numpy as np
x = pd.DataFrame({'x': np.linspace(-5, 5)})
```

### График синуса и косинуса с использованием `transform_calculate`

```python
base=alt.Chart(x).mark_line().encode(x='x:Q')
```

```python
sin_chart = base.mark_line(color='blue').encode(
    x='x:Q',
    y='sin:Q'
).properties(
    title='Sine Chart'
)
sin_chart.show()
```

```python
cos_chart = base.mark_line(color='red').encode(
    x='x:Q',
    y='cos:Q'
).properties(
    title='Cosine Chart'
)
cos_chart.show()
```

### `transform_filter` на графиках синуса и косинуса

```python
filtered_chart = (sin_chart + cos_chart).transform_filter(
    'datum.cos >= datum.sin'
)

final_chart = filtered_chart.properties(
    width=600,
    height=400,
    title='Combined Trigonometric Functions (cos ≥ sin)'
)

final_chart.show()
```
